```python
import yfinance as yf
import numpy as np
import pandas as pd

payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
first_table = payload[0]
second_table = payload[1]
#print(first_table.head())
sip=np.array(first_table.iloc[:,0])
df=pd.DataFrame()

yld=0.0412
#for i in range(0,len(sip)):
for i in range(0,1):
    #ticker=sip[i]
    ticker="TQQQ"
    start_date="2020-08-31"
    end_date="2022-06-08"
    intr="1wk"
    data = yf.download(ticker, start_date, end_date,interval=intr).dropna()
    data["ChangeP"]=(data.Close-data.Open)/data.Open
    data["Change"]=(data.Close-data.Open)
    data["Yield"]=(data.Open)*yld
    
    
    data.loc[data.Change > 0,"PutPL"]=data.Yield
    data.loc[data.Change < 0,"PutPL"]=data.Yield+data.Change
    
    data.loc[data.Change > 0,"CallPL"]=data.Yield-data.Change
    data.loc[data.Change < 0,"CallPL"]=data.Yield
   
    data["TotalPL"]=data.CallPL+data.PutPL

    Sigma=(np.std(data.ChangeP))
    AverageYield=np.average(data.TotalPL)
    TotalReturn=np.sum(data.TotalPL)
    WinRate=data[data.TotalPL>0].shape[0]/data.shape[0]
    #err=abs(STD-STDCC)
    #print(data.tail())
    #dict=dict.append({"tick":1,"STD%":2})
    #print(WIN)
    temp=pd.Series([ticker,Sigma,AverageYield,TotalReturn,WinRate])
    df=df.append(temp,ignore_index=True)
    
    #data.Change.apply(lambda x: data.Yield if x<0 else 0)
   # np.select(data.Change>0,data.Yield,data.Yield+data.Change)
print(df)
print(data.head)
data.to_csv('exit.csv')
```
