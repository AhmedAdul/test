```python
import yfinance as yf
import numpy as np
import pandas as pd

payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
first_table = payload[0]
second_table = payload[1]
#print(first_table.head())
sip=np.array(first_table.iloc[:,0])
df=pd.DataFrame()
for i in range(0,len(sip)):
#for i in range(0,20):
    ticker=sip[i]
    start_date="2010-01-01"
    end_date="2021-10-01"
    intr="1wk"
    data = yf.download(ticker, start_date, end_date,interval=intr).dropna()
    data["ChangeP"]=(data.Close-data.Open)/data.Open
    data["Close1"]=data.Close.shift(1)
    data["ChangeCC"]=(data.Close-data.Close1)/data.Close
    STD=(np.std(data.ChangeP))
    STDCC=(np.std(data.ChangeCC))
    err=abs(STD-STDCC)
    #print(data.tail())
    #dict=dict.append({"tick":1,"STD%":2})
    temp=pd.Series([ticker,STD,STDCC,err])
    df=df.append(temp,ignore_index=True)


df.columns=['stock','STD_open-close','STD_close-close','error']
    #print(data.head())
print(df)
avg_err="{:.2%}".format(np.average(df.error))
print(avg_err)
df.to_csv('spy_std.csv')
```

```python
df=df.dropna()
avg_err="{:.2%}".format(np.average(df.error))
max_err="{:.2%}".format(np.max(df.error))
avg_std1="{:.2%}".format(np.average(df["STD_open-close"]))
avg_std2="{:.2%}".format(np.average(df["STD_close-close"]))
print(avg_err,max_err,avg_std1,avg_std2)
```
