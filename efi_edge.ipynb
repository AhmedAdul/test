```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import date
from ta.momentum import RSIIndicator
from ta.volume import ForceIndexIndicator
payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
#print(first_table.head())
sip=np.array(payload.iloc[:,0])
df=pd.DataFrame()

start_date="2015-01-01"
end_date="2021-10-01"
intr="1wk"
dfout=pd.DataFrame()
for i in range(0,len(sip)):
#for i in range(0,20):
    stk=sip[i]
    df = yf.download(stk, start_date, end_date,interval=intr)
    df=df.dropna()
    
    df["ChngP"]=(df.Close-df.Open)/df.Open
    efi=ForceIndexIndicator(close=df.Close,volume=df.Volume,window=14,fillna=True)
    df['efi']=efi.force_index().shift(1)
    rsi=RSIIndicator(close=df.efi,window=14,fillna=True)
    df['efi_rsi']=rsi.rsi()
    df=df.iloc[40:10000,:]
    ovrs=df[df['efi_rsi']<40]
    ovrb=df[df['efi_rsi']>60]
    ovrs_chng=np.average(ovrs.ChngP)
    ovrb_chng=np.average(ovrb.ChngP)
    ovrs_edg=(ovrs_chng-np.average(df.ChngP))
    ovrb_edg=(ovrb_chng-np.average(df.ChngP))
    temp=pd.Series([stk,ovrs_edg,ovrb_edg])
    dfout=dfout.append(temp,ignore_index=True)
    
    
    
    #print(dfout.head())
dfout.columns=['stock','oversold_edge','overbought_edge']
print(dfout)
#dfout.to_csv('stochrsi_week_edge.csv')
```

```python
dfout=dfout.dropna()
print("{:.2%}".format(np.average(dfout.oversold_edge)))
print("{:.2%}".format(np.average(dfout.overbought_edge)))
```
