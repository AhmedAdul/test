```python
import yfinance as yf
import pandas as pd
import numpy as np


class main:
    
    def __init__(self,stock,start,end,interval):
        self.stock=stock
        self.start=start
        self.end=end
        self.interval=interval
    
    def get_data_yf(self):
        df=(yf.download(self.stock,self.start,self.end,interval=self.interval).dropna())
        return(df)
        
    def rounding(base,df):
        df=df.apply(lambda x: base*round(x/base))
        return(df)
    
    def ohlc_hit_cnt(df):
        dfo=df.groupby(['Open']).count().iloc[:,0].to_frame()
        dfc=df.groupby(['Close']).count().iloc[:,0].to_frame()
        dfh=df.groupby(['High']).count().iloc[:,0].to_frame()
        dfl=df.groupby(['Low']).count().iloc[:,0].to_frame()
        dff=dfo.join(dfc,rsuffix="o").join(dfh,rsuffix="o").join(dfl,rsuffix="o")
        dff.columns=['Open','Close','High','Low']
        dff['total']=dff.sum(axis=1)
        dff['Rank']=dff.total.rank(pct=True)
        return(dff)

    def vol_agg(df):
        dfo=df.groupby(['Open']).sum().Volume.to_frame()
        dfc=df.groupby(['Close']).sum().Volume.to_frame()
        dfh=df.groupby(['High']).sum().Volume.to_frame()
        dfl=df.groupby(['Low']).sum().Volume.to_frame()
        dff=dfo.join(dfc,rsuffix="o").join(dfh,rsuffix="o").join(dfl,rsuffix="o")
        dff.columns=['Open','Close','High','Low']
        dff['total']=dff.sum(axis=1)
        dff['Rank']=dff.total.rank(pct=True)
        
        return(dff)
    
start="2000-01-04"
end='2025-12-04'
interval='1d'
stock="T"
base=1
                    
ini=main(stock,start,end,interval)

rawdata=ini.get_data_yf()

data=main.rounding(base,rawdata)

ohlc=main.ohlc_hit_cnt(data)

print(main.vol_agg(data))

print(ohlc)
```
