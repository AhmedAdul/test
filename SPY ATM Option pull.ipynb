```python

import yfinance as yf
payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
#print(first_table.head())
sip=np.array(payload.iloc[:,0])
df=pd.DataFrame()
for i in range(0,len(sip)):
    try:
        stk=sip[i]
        tick= yf.Ticker(stk)
        ch=tick.option_chain(date='2021-10-22').calls
        ch=ch[ch.inTheMoney==True]
        ln=len(ch.iloc[:,0])
        yld=(ch.iloc[ln-1,3])/ch.iloc[ln-1,2]
        strike=ch.iloc[ln-1,2]
        premium=ch.iloc[ln-1,3]
        temp=pd.Series([stk,strike,premium,yld])
        df=df.append(temp,ignore_index=True)
    except:
        pass
print(df)
df.to_csv("yield.csv")
```
