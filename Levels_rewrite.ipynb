```python
import yfinance as yf
import pandas as pd
import numpy as np


class main:
    
    def __init__(self,stock,start,end,interval):
        self.stock=stock
        self.start=start
        self.end=end
        self.interval=interval
    
    def get_data_yf(self):
        df=(yf.download(self.stock,self.start,self.end,interval=self.interval).dropna())
        return(df)
        
    def rounding(base,df):
        df=df.apply(lambda x: base*round(x/base))
        return(df)
    
    def ohlc_hit_cnt(df):
        dfc=df.groupby(['Close']).count().iloc[:,0]
        dfo=df.groupby(['Open']).count().iloc[:,0]
        dfh=df.groupby(['High']).count().iloc[:,0]
        dfl=df.groupby(['Low']).count().iloc[:,0]
        dff={'Close':dfc,'Open':dfo,'High':dfh}
        #dff['total']=dff.sum(axis=1)
        return(dff)

start="2000-01-04"
end='2025-12-04'
interval='1d'
stock="T"
base=1
                    
ini=main(stock,start,end,interval)

rawdata=ini.get_data_yf()

data=main.rounding(base,rawdata)

data=main.ohlc_hit_cnt(data)

print(data)
```
