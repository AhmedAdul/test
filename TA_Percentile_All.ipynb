```python
import yfinance as yf
import pandas as pd
import numpy as np
from ta import add_all_ta_features
payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
#print(first_table.head())
sip=np.array(payload.iloc[:,0])
sip=sip[sip!='BRK.B']
sip=sip[sip!='BF.B']
df1=pd.DataFrame()
df2=pd.DataFrame()
dfout=pd.DataFrame()
start_date="2000-10-01"
end_date="2021-10-01"
txt = "{:%}"
wind=14
for i in range(0,len(sip)):
    try:
        stk=sip[i]
        #df = yf.download(stk, start_date, end_date,interval=intr).dropna()

        dfd = yf.download(stk, start_date, end_date,interval='1d').dropna()
        dfd['Open1']=dfd.Open.shift(1)
        dfd['Close1']=dfd.Close.shift(1)
        dfd['High1']=dfd.High.shift(1)
        dfd['Low1']=dfd.Low.shift(1)
        dfd['Volume1']=dfd.Volume.shift(1)
        #dfd['ChangePre']=dfd.Close-dfd.Close1
        dfd['Change']=dfd.Close-dfd.Open
        #dfd['ChangePreP']=dfd.ChangePre/dfd.Close1
        dfd['ChangeP']=dfd.Change/dfd.Open
        fir=dfd.iloc[0,0]
        las=dfd.iloc[len(dfd.index)-1,3]
        dfd = add_all_ta_features(
        dfd, open="Open1", high="High1", low="Low1", close="Close1", volume="Volume1", fillna=True)
        dfd=dfd.dropna()
        temp=np.array([])
        for j in range(13,len(dfd.columns)):
            #print(dfd.columns[j])
            perc=np.percentile(dfd.iloc[:,j],1)
            dft=dfd[dfd.iloc[:,j]<perc]
            #print(perc)
            #print(dft.tail)
            ch=(np.average(dft.ChangeP))
            temp=np.append(temp,ch)

        df1=pd.DataFrame(temp.reshape(1,-1))
        df1.index=([stk])
        df2=pd.concat([df2,df1])
        
    except:
        print(Exception)
        pass
df2.columns=([dfd.columns[13:len(dfd.columns)]])
print(df2)

fin=np.array([])
for k in range (0,len(df2.columns)):
    temp1=(np.average(df2.iloc[:,k]))    
    fin=np.append(fin,temp1)
dff=pd.DataFrame(fin)   

dff.index=df2.columns
print(dff)
dff.to_csv('ta_al.csv')
```
