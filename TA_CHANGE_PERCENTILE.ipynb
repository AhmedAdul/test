```python
import sys
!{sys.executable} -m pip install scipy

import yfinance as yf
import pandas as pd
import numpy as np
from scipy import stats
from ta import add_all_ta_features
from ta.utils import dropna


# Add ta features filling NaN values


#from scipy.stats import percentileofscore
payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
#print(first_table.head())
sip=np.array(payload.iloc[:,0])
sip=sip[sip!='BRK.B']
sip=sip[sip!='BF.B']


start_date="2000-10-01"
end_date="2021-10-01"
txt = "{:.2%}"

dates=['1d','1wk','1mo']
for t in range(0,1):
    stklst=np.array([])
    temp2=np.array([])
    for i in range(0,len(sip)):
        try:
            stk=sip[i]

            dfd = yf.download(stk, start_date, end_date,interval=dates[t]).dropna()
            dfd['Open1']=dfd.Open.shift(1)
            dfd['Close1']=dfd.Close.shift(1)
            dfd['High1']=dfd.High.shift(1)
            dfd['Low1']=dfd.Low.shift(1)
            dfd['Volume1']=dfd.Volume.shift(1)
            dfd['Change']=(dfd.Close-dfd.Open)/dfd.Open


            dfd = add_all_ta_features(
            dfd, open="Open1", high="High1", low="Low1", close="Close1", volume="Volume1", fillna=True).dropna()


            dfdtop=dfd[dfd['Change']<=np.percentile(dfd.Change,1)]

            temp=np.array([])
            for j in range (12,len(dfd.columns)):
                ave=np.nanmean(dfdtop.iloc[:,j])
                perc=stats.percentileofscore(dfd.iloc[:,j],ave)
                temp=np.append(temp,perc)

            temp2=np.append(temp2,temp)
            #print(temp2)
            stklst=np.append(stklst,stk)
        except Exception as e: print(e)
    temp2=temp2.reshape(len(stklst),j-11)
    #print(temp2)
    dff=pd.DataFrame(temp2,index=[stklst],columns=[dfd.columns[12:len(dfd.columns)]])
    #print(dff)
    dff.to_csv(F'ta_percentile_raw_{dates[t]}.csv ')
    
    
    temp_3=np.array([])
    for k in range (0,len(dff.columns)):
        mean_3=(np.nanmean(dff.iloc[:,k]))    
        temp_3=np.append(mean_3,temp_3)
    #print(dff.columns)
    dff_2=pd.DataFrame(temp_3)
    dff_2.set_index(dff.columns,inplace=True)
    dff_2.sort_values(by=[0],axis=0,ascending=False,inplace=True)
    #print(dff_2)
    dff_2.to_csv(F'ta_percentile_clean_{dates[t]}.csv ')
    
    
    
   
```
