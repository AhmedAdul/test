```python
import sys

!{sys.executable} -m pip install Levenshtein
import Levenshtein

import pandas as pd
df=pd.read_csv('job_skills.csv')
#print(df.head)
txt=df['Preferred Qualifications'].tolist()
#print(txt)


def SplitLines(df):
    
    SplitLineList=[]
    
    for job in df:
        
        try:
            lines=(job.splitlines())
        except: 
            pass
        
        for line in lines:
                SplitLineList.append(line)
                
    return(SplitLineList)


def SplitComas(df):

    SplitLineList=[]
    
    for job in df:
        
        try:
            lines=(job.split("."))
        except: 
            pass
        
        for line in lines:
                SplitLineList.append(line)
                
    return(SplitLineList)

def CleanSimiliarLevenshtein(StringList):
    for i in range(0,len(StringList)):
        for j in range(0,len(StringList)):
            adj_score=Levenshtein.ratio(StringList[i],StringList[j])
            #print(adj_score)
            if adj_score>0.8:
                StringList[j]=StringList[i]
    return(StringList)


def GroupByFrequency(df):
    df_freq=df.groupby(['lines'])['lines'].count()
    return(df_freq)        
    

temp=SplitLines(txt)
#temp=SplitLines(txt)[0:1200]
temp=CleanSimiliarLevenshtein(temp)
#print(temp)
df1=pd.DataFrame(temp,columns=['lines'])
#print(df1)
df1.to_csv('splitlines.csv')
df2=GroupByFrequency(df1).sort_values(ascending=False)

print(df2)
df2.to_csv('groups3.csv')

```
