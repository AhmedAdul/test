```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('DMR.csv',encoding = "ISO-8859-1")

#print(df.iloc[1,:])

jobs=[100,90,85,60,90,100,90,110,100,90]




#format time & set index
df['Sys Date']=pd.to_datetime(df['Sys Date'])
df['Sys Date']=df['Sys Date'].dt.strftime('%m')
df=df.set_index("Sys Date")
col_len=len(df.iloc[0,:])
df['DefGroup']=df.iloc[:,col_len-1].str[:3]
df['Class']=df.iloc[:,col_len-2]
#print(df.tail())
#Class aggregate
df['Class']= np.where((df['Class']=="INDUSTRIAL:Metal Fittings"), "Fittings", df['Class'])
df['Class']= np.where((df['Class']=="OTHER:Metal Fittings"), "Fittings", df['Class'])
df['Class']= np.where((df['Class']=="AERO:Metal Fittings"), "Fittings", df['Class'])

df['Class']= np.where((df['Class']=="AM-PM:Finished Item-Medical"), "AM-PM", df['Class'])
df['Class']= np.where((df['Class']=="AM-PM:Comp. Medical: FL"), "AM-PM", df['Class'])

df['Class']= np.where((df['Class']=="INDUSTRIAL:Finished Assy"), "Industrial", df['Class'])
df['Class']= np.where((df['Class']=="READYFLEX:Finished Assy"), "Industrial", df['Class'])

df['Class']= np.where((df['Class']=="AERO:Finished Assy"), "Aerospace", df['Class'])

df['Class']= np.where((df['Class']=="MEDICAL:Finished Assy"), "Flex Med", df['Class'])

#Class Match column & set rows to 0
cls="Aerospace"
df['TrType']=df.iloc[:,11]
df1=df[df['Class']==cls]
df1=df1[df1['TrType']=="I"]
#df["chk12"]=df['Class']==cls
#df.loc[df["chk12"]==False]=0


#df1=df
#df1["chk11"]=(df1['TrType']=="I") | (df1['TrType']=="M") 
#df2=df1.iloc[df1["chk11"]==False]=0
#print(df1)

#pivot table
p1=pd.pivot_table(df1,index='Sys Date',columns='DefGroup',values='DMR Number',aggfunc='count')
p1=p1.fillna(0)
#p1=p1.drop(columns=[0])
#print(p1)
hor_len=len(p1.iloc[0,:].to_numpy())
ver_len=len(p1.iloc[:,0].to_numpy())


zeros=np.zeros((12,4))
slate=pd.DataFrame(zeros,index=['01','02','03','04','05','06','07','08','09','10','11','12'],columns=p1.columns().to_numpy())





bot=[0]*ver_len
lbl=p1.columns.to_numpy()
ind = np.arange(ver_len)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(3,figsize=(16,9))

for i in range(0,hor_len):
    y=p1.iloc[:,i].to_numpy()
    p=ax[0].bar(ind,y,width,bottom=bot,label=lbl[i])
    #print(y)
    #print(x)
    #plt.bar(x,y,bottom=bot,label=lbl[i])
    bot=np.add(bot,y)
    ax[0].bar_label(p,label_type='center')
    
#p1 = ax.bar(ind, menMeans, width, label='Men')
#p2 = ax.bar(ind, womenMeans, width,bottom=menMeans, label='Women')

ax[0].axhline(0, color='grey', linewidth=0.8)
ax[0].set_ylabel("DMR count")
ax[0].set_title(cls+" DMR Defect Count")
ax[0].set_xlabel("Month")
ax[0].set_xticks(ind)
ax[0].set_xticklabels(('JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT'))
ax[0].set_ylim(0,np.max(bot)+5)
ax[0].legend()
# Label with label_type 'center' instead of the default 'edge'
#ax.bar_label(p1, label_type='center')
#ax.bar_label(p2, label_type='center')
#ax.bar_label(p2)






plt.show()
```
